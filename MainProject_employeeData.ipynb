##### Filling Missing Age Value
To handle the missing age value for one employee, the age was recalculated using the following formula:

1. Retrieved the employee’s Date of Birth (DOB).
2. Subtracted the DOB from the current date using Pandas Timestamp.
3. Converted the days difference into years by dividing by 365.25.
4. Rounded the result to the nearest whole number.
5. Updated the `Age` column for the employee whose `EmployeeID` was 1475.

```python
today = pd.Timestamp.today()
emp_id = 1475

df.loc[df['EmployeeID'] == emp_id, 'Age'] = (
    (today - df.loc[df['EmployeeID'] == emp_id, 'DOB'])
    .dt.days / 365.25
).round(0)
```

```python
age_value = df.loc[df['EmployeeID'] == 1475, 'Age'].values[0]
print("Updated Age =", age_value)
```

##### Handling Missing Values in `ManagerID`

```python
df['LastPromotionDate'] = df['LastPromotionDate'].fillna(pd.NaT)
```

##### Data Type Correction
To ensure the dataset is clean, consistent, and optimized for analysis, several datatype corrections were performed:
######   Numeric Columns Converted to Integer
Columns such as `EmployeeID`, `SalaryINR`, `Phone`, `PerformanceScore`, and `WorkExperienceYears` were converted to `int` because they represent numerical identifiers or counts.
######  ManagerID Converted to Integer
After replacing missing values with `0`, `ManagerID` was converted from `float` to `int`, ensuring a clean reporting structure.
######  Age Converted to Integer
The `Age` column originally contained floating values due to calculations.  
After imputing the missing age, the column was converted to integer for consistency.
######  Date Columns Converted to Datetime
`DOB`, `HireDate`, and `LastPromotionDate` were converted to `datetime64[ns]` format to enable:
- Time-based calculations  
- Duration analysis  
- Sorting and filtering by date  
Invalid or missing promotion dates were automatically converted to `NaT`.
######  Categorical Columns Converted to Category Type
Several string-based columns such as `Gender`, `Department`, `JobTitle`, `Location`, `Education`, etc., were converted to `category` datatype to:
- Improve processing speed  
- Maintain consistency in categorical fields  
This ensures efficient modeling and grouping operations.


```python
# Convert IDs and numeric fields to integer
df['EmployeeID'] = df['EmployeeID'].astype(int)
df['SalaryINR'] = df['SalaryINR'].astype(int)
df['Phone'] = df['Phone'].astype(int)
df['PerformanceScore'] = df['PerformanceScore'].astype(int)
df['WorkExperienceYears'] = df['WorkExperienceYears'].astype(int)

# Convert ManagerID (missing values already filled with 0)
df['ManagerID'] = df['ManagerID'].astype(int)

# Convert Age (after filling missing value)
df['Age'] = df['Age'].astype(int)

# Convert categorical text fields to category datatype to save memory
categorical_cols = [
    'FirstName', 'LastName', 'Gender', 'Department', 'JobTitle', 'Location',
    'Email', 'Education', 'MaritalStatus', 'EmploymentType', 'Active'
]
df[categorical_cols] = df[categorical_cols].astype('category')

# Ensure date fields are in datetime format
df['DOB'] = pd.to_datetime(df['DOB'])
df['HireDate'] = pd.to_datetime(df['HireDate'])
df['LastPromotionDate'] = pd.to_datetime(df['LastPromotionDate'], errors='coerce')
```

```python
df.info()
```

##### 5. Creation of Derived Columns
To enrich the dataset and support deeper analysis, several new calculated fields were added.

```python
# 1. Tenure in Years
df['TenureYears'] = ((pd.to_datetime("today") - df['HireDate']).dt.days / 365).round(2)

# 2. Age Group
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[18, 30, 40, 50, 60],
    labels=['18-30', '31-40', '41-50', '51-60'])

# 3. Full Name
df['EmployeeName'] = df['FirstName'].astype(str) + " " + df['LastName'].astype(str)

# 4. Monthly Salary + LPA (Lakhs Per Annum)
df['Salary_Monthly'] = df['SalaryINR'] / 12
df['Salary_LPA'] = (df['SalaryINR'] / 100000).round(2)

# 5. Convert Performance Score to Integer
df['PerformanceScore'] = df['PerformanceScore'].astype(int)

# 6. Performance Rating Label
def performance_rating(score):
    if score == 5:
        return 'Excellent'
    elif score == 4:
        return 'Very Good'
    elif score == 3:
        return 'Good'
    elif score == 2:
        return 'Average'
    else:
        return 'Poor'

df['PerformanceRating'] = df['PerformanceScore'].apply(performance_rating)
# 7. 
df['SalaryCategory'] = pd.cut(
    df['SalaryINR'],
    bins=[0, 300000, 500000, 900000, 1500000],
    labels=['Low', 'Medium', 'High', 'Very High'],
    include_lowest=True
)

```
